```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import plotly.express as px
```

```python
df = df[df['Average Purchase Value']>=0]
df = df[df['Total Values of Returns']>=0]
df = df[df['Last Purchase Value']>=0]
df.shape
```

```python
df['total_purchase'] = df['Number of Online Purchases'] * df['Average Purchase Value'] - df['Total Values of Returns']
```

```python
fig = px.box(df, y="total_purchase", points="outliers")
fig.show()
```

```python
fig = px.box(df,x='Gender', y="Income", points="outliers")
fig.show()
```

```python
fig = px.histogram(df,
                  x='total_purchase',
                  title ='Card Limit',
                  marginal = 'box',
                  color = 'Occupation')
fig.update_layout(bargap = 0.1)
fig.show()
```
